## Dictionary

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 21,
    "course": "Computer Science",
    "grades": [85, 90, 78]
}

# Accessing values
print(student["name"])       # Output: Alice
print(student.get("age"))    # Output: 21

# Adding new key-value pairs
student["university"] = "MIT"

# Updating values
student["age"] = 22

# Removing items
student.pop("course")        # Removes 'course'
del student["grades"]        # Removes 'grades'

# Iterating through dictionary
for key, value in student.items():
    print(key, ":", value)

# Checking if a key exists
if "name" in student:
    print("Name exists in dictionary")

# Dictionary methods
print(student.keys())        # Returns all keys
print(student.values())      # Returns all values
print(student.items())       # Returns all key-value pairs

# Copying dictionary
student_copy = student.copy()

# Clearing dictionary
student.clear()
print(student)               # Output: {}
```

pop - Removes the item with the specified key     
popitems - Removes the last inserted key-value pair

Nested Dictionary

```python
students = {
    "student1": {
        "name": "Alice",
        "age": 21,
        "course": "Computer Science"
    },
    "student2": {
        "name": "Bob",
        "age": 22,
        "course": "Mechanical Engineering"
    },
    "student3": {
        "name": "Charlie",
        "age": 20,
        "course": "Mathematics"
    }
}

# Accessing nested values
print(students["student1"]["name"])   
print(students["student2"]["course"],f'\n') 

# Updating nested dictionary
students["student3"]["age"] = 21

# Adding new nested entry
students["student4"] = {
    "name": "Diana",
    "age": 23,
    "course": "Physics"
}

# Iterating through nested dictionary
for student_id, details in students.items():
    print(student_id, ":", details)
```

Separating even and odd from a list

```python

numbers = [10, 21, 4, 45, 66, 93, 11]
evens = []
odds = []

for num in numbers:
    if num % 2 == 0:
        evens.append(num)
    else:
        odds.append(num)
print("Even numbers:", evens)
print("Odd numbers:", odds)
```

```python
numbers = [10, 21, 4, 45, 66, 93, 11]
for i in numbers:
    if i % 2 == 0:
        print(i," - even")
    else:
        print(i," - odd")
```

Separating int,float,str from list

```python
l=[10,42,67,3.3,4.2,4.0,'s','o']
type_String=[]
type_int=[]
type_float=[]
for i in l:
    if type(i)==int:
        type_int.append(i)
    elif type(i)==float:
        type_float.append(i)
    else:
        type_String.append(i)
print(type_int)
print(type_float)
print(type_String)
```

check for palindrome

```python
def is_palindrome():
    num = int(input("Enter a number: "))
    if num == int(str(num)[::-1]):
        print(f"{num} is a palindrome.")
    else:
        print(f"{num} is not a palindrome.")

is_palindrome()
```

sum of entered digit

```python
def sum_of_digit():
  num = input("Enter a number: ")
  digit_sum = 0

  for ch in num:
    digit_sum += int(ch)   

  print("Sum of digits:", digit_sum)

sum_of_digit()
```
